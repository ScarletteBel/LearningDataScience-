```python
import sklearn 
import pandas as pd 
from tabulate import tabulate
from sklearn import datasets 
from sklearn import svm 
```

```python
diabetes = datasets.load_diabetes()
#print(diabetes)
```

```python
columns = diabetes.feature_names
print(columns)
print(tabulate(diabetes.data))
```

```python
x = diabetes.data
y = diabetes.target
```

```python
xshape = x.shape
yshape = y.shape 
```

```python
print(xshape)
print(yshape)
```

Now...let's go to split the data 

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.20)
```

```python
print()
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

let's go to model the linear regression!!!

```python
from sklearn import linear_model 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
model = linear_model.LinearRegression()
trainmodel = model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

Linear regression model performance verifiers...

```python
print()
print("Coeficients:",model.coef_)
print("Intercept:",model.intercept_)
print("Mean squared error (MSE): ", mean_squared_error(y_test,y_pred))
print("Coefficient of determination (R^2): ", r2_score(y_test,y_pred))
```
